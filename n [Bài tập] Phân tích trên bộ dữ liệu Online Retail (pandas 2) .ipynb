### 1. Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính

```python
import pandas as pd
import numpy as np
df = pd.read_csv('OnlineRetail.csv', encoding = 'ISO-8859-1')
df
```

```python
df.info()
```

+ InvoiceNo, StockCode, Description, InvoiceDate, CustomerID, Country: thang đo định danh
+ Quantity, Unit Price: thang đo tỉ lệ

### 2. Xây dựng bảng Pivot Table, với mỗi Số hoá đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia

```python
# đổi tên cột sang tiếng việt
df.rename(columns = {'InvoiceNo':'Sohoadon','StockCode':'Macophieu','Description':'Mota','InvoiceDate':'Ngay','CustomerID':'Makhachhang','Country':'Nuoc','Quantity':'Soluong','UnitPrice':'Gia'},inplace = True)
df.head()
```

```python
df.pivot_table(values = 'Soluong', index = 'Sohoadon', columns = 'Nuoc', aggfunc = 'mean')
```

### 3. Xây dựng bảng Pivot Table, với mỗi khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo Kho

```python
df.pivot_table(values = 'Soluong', index = 'Macophieu', columns = 'Makhachhang', aggfunc = {min,max})
```

### 4. Xây dựng bảng Pivot table, với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá

```python
df.pivot_table(values = ['Soluong','Gia'], index = 'Macophieu', aggfunc = {'Soluong':np.sum,'Gia':np.mean})
```

### 5. Xây dựng bảng Pivot table cho biết tổng số lượng hàng bán được của mỗi ngày

```python
# lọc ngày
a = [x[:9] for x in df['Ngay']]
df['Ngay']= a
df.head()
```

```python
df.pivot_table(values = 'Soluong', index = 'Ngay', aggfunc = 'sum')
```
